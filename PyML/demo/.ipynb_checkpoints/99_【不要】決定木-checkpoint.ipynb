決定木

# ライブラリのインポート

```python
import numpy as np
import pandas as pd
import sklearn.tree as tree
from IPython.display import Image
import pydotplus
from sklearn.externals.six import StringIO
```

# ファイルの読み込み

```python
data = pd.read_csv('01_LiquidRank_2値.csv',encoding='shift_jis')
data.head()
```

# 決定木を呼び出し

```python
#clf = tree.DecisionTreeClassifier() #デフォルトだと、階層が深くなりすぎる
#clf = tree.DecisionTreeClassifier(max_depth=2)
clf = tree.DecisionTreeClassifier(max_depth=2)
```

# 決定木の適用

```python
clf = clf.fit(X=data.ix[:,0:11],y=data['ランク'])
```

# 決定木の評価

重要度の算出

```python
clf.feature_importances_
```

正解率の算出

```python
clf.score(X=data.ix[:,0:11],y=data['ランク'])
```

# モデルの確認（可視化）

可視化に使うデータを用意

```python
features=data.ix[:,0:11].columns #2列目から8列目までの列名を変数に格納
targets=np.array(data['ランク']) # 目的変数の値を変数に格納
targets=np.unique(targets) # 変数の中身を重複排除
```

可視化に使うルールを、テキスト（dot形式）として変数に格納

```python
dot_tmp=StringIO()
tree.export_graphviz(clf, out_file=dot_tmp,
                                feature_names=features,
                                class_names=targets,
                               filled=True, rounded=True,
                               special_characters=True)  
```

テキスト(dot形式)からグラフ化し、pngイメージ化

```python
graph = pydotplus.graph_from_dot_data(dot_tmp.getvalue())
Image(graph.create_png())
```

# 特徴絞って決定木

```python
#clf = tree.DecisionTreeClassifier() #デフォルトだと、階層が深くなりすぎる
clf = tree.DecisionTreeClassifier(max_depth=2)
```

# 決定木の適用

```python
features=[-5,-4,-2]
clf = clf.fit(X=data.ix[:,features],y=data['ランク'])
```
