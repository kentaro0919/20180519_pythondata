### 5. 以下の情報が確認できるアウトプットを、表などで作成してください。

店舗名別のデータの個数（注文番号数）

```python
set(df["店舗名"])
```

```python
print(
    [(d, len(df[df["店舗名"]==d]))for d in set(df["店舗名"]) if d == '01A'],
    [(d, len(df[df["店舗名"]==d]))for d in set(df["店舗名"]) if d == '02B'],
    [(d, len(df[df["店舗名"]==d]))for d in set(df["店舗名"]) if d == '03C'],
     )
```

```python
df["店舗名"].value_counts()
```

「店舗名×来店時間帯」別のデータの個数（注文番号数）

```python
df["店舗名"].value_counts()
```

「店舗名×来店時間帯」別の滞留時間の平均

```python
aa = df.groupby("店舗名")["滞留時間"].mean()
aa
```

「店舗名×来店時間帯」別の合計金額の合計

```python
bb = df.groupby("店舗名")["滞留時間"].sum()
bb
```
